# Reading and Parsing Data

```python
filename="input.txt"
elves=[]
with open(filename) as f:
    elves = f.read().strip().split("\n\n")
    elves = list(map(lambda x: x.replace("\n", " ").split(" "), elves))
```

```python
elf_totals = list(map(lambda elf: sum(list(map(lambda calories: int(calories),elf))), elves))
sorted_elf_totals = sorted(elf_totals)
print(sorted_elf_totals)
```
